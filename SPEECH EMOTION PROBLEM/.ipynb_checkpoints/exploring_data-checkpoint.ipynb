Plotting a Spectrogram using Scipy library:
A spectrogram is a visual representation of the spectrum of frequencies as it varies with time. DFT (Discrete Fourier transform) is used to compute the same. There exist various types of windowing functions in DFT, in particular, the Hanning window is used here.

http://practicalcryptography.com/miscellaneous/machine-learning/guide-mel-frequency-cepstral-coefficients-mfccs/
MFFC the Kaggle post mentions it so I searched about it. However, after searching about I found that it is useful for extracting features for speech recognition only and it removes emotional features so most probably it will not be useful in our case.


```python
def log_specgram(audio, sample_rate, window_size=20,
                 step_size=10, eps=1e-10):
    nperseg = int(round(window_size * sample_rate / 1e3))
    noverlap = int(round(step_size * sample_rate / 1e3))
    freqs, times, spec = signal.spectrogram(audio,
                                    fs=sample_rate,
                                    window='hann',
                                   # nperseg=nperseg,
                                   # noverlap=noverlap,
                                    detrend=False)
    return freqs, times, np.log(spec.T.astype(np.float32) + eps)
```

```python
freqs, times, spectrogram = log_specgram(channel1, sample_rate)

fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1.set_title('Raw wave of channel 1 ' + filename)
ax1.set_ylabel('Amplitude')
ax1.plot(np.linspace(0, sample_rate/len(samples), len(samples)), channel1)

ax2 = fig.add_subplot(212)
ax2.imshow(np.array(spectrogram.T), aspect='auto', origin='lower', 
           extent=[times.min(), times.max(), freqs.min(), freqs.max()])
ax2.set_yticks(freqs[::16])
ax2.set_xticks(times[::16])
ax2.set_title('Spectrogram of ' + filename)
ax2.set_ylabel('Freqs in Hz')
ax2.set_xlabel('Seconds')
```

```python
freqs, times, spectrogram = log_specgram(channel2, sample_rate)

fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1.set_title('Raw wave of channel 1 ' + filename)
ax1.set_ylabel('Amplitude')
ax1.plot(np.linspace(0, sample_rate/len(samples), len(samples)), channel2)

ax2 = fig.add_subplot(212)
ax2.imshow(np.array(spectrogram.T), aspect='auto', origin='lower', 
           extent=[times.min(), times.max(), freqs.min(), freqs.max()])
ax2.set_yticks(freqs[::16])
ax2.set_xticks(times[::16])
ax2.set_title('Spectrogram of ' + filename)
ax2.set_ylabel('Freqs in Hz')
ax2.set_xlabel('Seconds')
```
