```python
import os
import random
from tqdm import tqdm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from python_speech_features import mfcc, logfbank
import librosa
import librosa.display
```

exploring the features in librosa

```python
train_audio_path = 'meld/train/happy'
categ=os.listdir(train_audio_path)
```

```python
print(categ)
```

```python
filename = '/MEL_dia108_utt5_negative_HAP.wav'
sample_rate, samples = wavfile.read(str(train_audio_path) + filename)
```

```python
samples, sample_rate = librosa.load(str(train_audio_path)+filename)
```

```python
tonnetz = librosa.feature.tonnetz(y=samples, sr=sample_rate)
```

```python
librosa.display.specshow(tonnetz, y_axis='tonnetz')
plt.colorbar()
plt.title('Tonal Centroids (Tonnetz)')
```

```python
chroma = librosa.feature.chroma_stft(y=samples, sr=sample_rate)
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chromagram')
plt.tight_layout()
plt.show()
```
